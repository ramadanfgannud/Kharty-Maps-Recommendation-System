```python
#Collaborative Filtering
#similarity between users
from scipy import stats
import numpy as np
import operator

def subtract(a,b):
    result = [] 
    for i in range(0, len(a)): 
        result.append(a[i] - b[i]) 
    return result

def CountFrequency(my_list): 
    freq = {} 
    for item in my_list: 
        if (item in freq): 
            freq[item] += 1
        else: 
            freq[item] = 1
    return freq

def corr(u, nu, nearestN, prnt):
    pearsons = {}
    for i in range(len(nu)):
        
        rownu = nu.iloc[i].tolist()
        avgnu = 0
        t = 0

        for d in rownu:
            if d > 0:
                avgnu += d
                t += 1
        if t != 0:
            avgnu = avgnu/t
        else:
            pass           #Can't recommend to this user (we can add this user to pearsons dict with value None)
        
        pearson = {}
        for n in range(len(u)):
            rowu = u.iloc[n].tolist()
            avgu = 0    #average of all maps rated
            tu = 0
            if rowu == rownu:    #skip user with itself
                continue
                
            for du in rowu:
                if du > 0:
                    avgu += du
                    tu += 1

            if tu != 0:
                avgu = avgu/tu
            else:
                continue

            user = []
            newuser = []
            for j in range(len(rownu)):
                if nu.iloc[i][j] > 0 and u.iloc[n][j] > 0:
                    user.append(u.iloc[n][j])
                    newuser.append(nu.iloc[i][j])
            if len(user) == 0:
                continue
            
            '''
            a = 0
            b = 0
            c = 0
            #newuser i and user n
            for k in range(len(user)):
                a += (user[k]-avgu)*(newuser[k]-avgnu)
                b += (user[k]-avgu)**2
                c += (newuser[k]-avgnu)**2

            p = a/((b*c)**0.5)
            #p,_ = stats.pearsonr(np.array(user),np.array(newuser))
            '''
            #pearson for similar user,newuser ratings will return Nan, Therefore,
            #Ratings should differ between the user and the new user to account for pearson corr
            
            p = len(user)     #This accounts for number of downloaded and created maps
            pearson[n] = p
        
        if len(pearson) > 0:
            if prnt == 1:
                print('Indexes of all users and their number of mutual downloads with target user {}:'.format(nu.index[i]))
                print(pearson,'\n')
            pearson = sorted(pearson, key=pearson.get, reverse=True)
            pearson = pearson[:nearestN]
            pearsons[i] = pearson 
    
    if prnt == 1:
        print('Indexes of most {} similar users to each of the following target users:\n'.format(nearestN))
        for newuserindex in pearsons:
            print('{} : {}'.format(nu.index[newuserindex], pearsons[newuserindex]))
    
    return pearsons
    

def recommend(df, u, nu, nearestN, prnt):
    pearsons = corr(u, nu, nearestN, prnt)
    uidx = u.index
    nuidx = nu.index
    index = []
    recommendations = []
    for i in pearsons:
        index.append(nu.index[i])
        rownu = nu.iloc[i]
        recommend = []
        for j in range(len(pearsons[i])):
            rowu = u.iloc[j]
            for k in range(len(rowu)):
                if rowu[k] != 0 and rownu[k] == 0:   #maps the new user didn't download yet a simmilar user downloaded
                    recommend.append(df.loc[df['objectId'] == u.columns[k], 'title'].iloc[0]) 

        recommend = CountFrequency(recommend)    #dictionary of maps with their number of repeatance within nearestN
        sorted_recommend = dict(sorted(recommend.items(), key=operator.itemgetter(1),reverse=True))
        recommendations.append([sorted_recommend])
    
    #create a pandas data frame for recall
    recommendations = pd.DataFrame(recommendations, index = index, columns= ['recommendedMaps'])    
    
    return recommendations

#Test example that takes first 100 users and all maps
#This takes first 10 users and recommend to them using most similar users in the 90 users left 
recommended = recommend(df, userdownloads.iloc[10:100,:],userdownloads.iloc[:10,:], 7, 1)
recommended
```

```python
#MAE using 0.8 tarin-test split for 10 target users based on 490 training users 
target_users = userdownloads.iloc[:10,:]     #I used first 10 users as targets
other_users = userdownloads.iloc[10:500,:]  #I used 490 users to recommend from
nearestN = 5
split = 0.8

def MAE(other_users, target_users, nearestN, split):
    
    train_columns = int(len(other_users.columns)*split)   
    test_columns = int(len(other_users.columns) - train_columns)  

    target_train_users = target_users.iloc[:,:train_columns]    
    other_train_users = other_users.iloc[:,:train_columns]  
    
    target_test_users = target_users.iloc[:,train_columns:]    
    other_test_users = other_users.iloc[:,train_columns:]
    
    target_users_index = target_users.index
    other_users_index = other_users.index
    
    train_corr = corr(other_users,target_users,nearestN,0)

    MAE = 0
    counter = 0
    for i in range(len(target_users)):    #iterate through target users
        if i in train_corr.keys():
            row_test_target = list(target_test_users.iloc[i])    #actual data 
            for j in train_corr[i]:
                row_test_users = list(other_test_users.iloc[j])
                MAE += (test_columns - CountFrequency(subtract(row_test_users, row_test_target))[0])
                counter += test_columns 
                
        else:    #That means this user wasn't returned for recommendation above because it has no download behaviors
            MAE += (test_columns - CountFrequency(subtract([0]*test_columns, row_test_target))[0])
            counter += test_columns 
    
    avg_MAE = MAE/counter 
    return avg_MAE
    

MAE(other_users, target_users, nearestN, split)
    
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer(r'\w+')
stop_words = nltk.corpus.stopwords.words()
stop_words = unique(stopWords['es'] + stop_words + stopWords['fr'])

def clear(sentence):
    words = tokenizer.tokenize(sentence)
    filtered_words = [w for w in words if w not in stop_words]
    ps = PorterStemmer()
    stemmed = [ps.stem(w) for w in filtered_words]
    return stemmed


print(df1['names'][0],'\n')
print(clear(df1['names'][0]))
```

```python
#get all unique words (columns)
columns = []
for i in range(len(df1)):
    df1.iloc[i]['names'] = clear(df1.iloc[i]['names'])
    for w in df1.iloc[i]['names']:
        if w not in columns:
            columns.append(w)

df1
```

```python
#########################################
#All maps and terms
#Maps-terms matrix

def TermFrequencies(column):
    rows = []
    for i in range(len(column)):
        row = []
        cell = column[i]
        for w in columns:
            if w in cell:
                row.append(cell.count(w))
            else:
                row.append(0)
        rows.append(row)
    
    return rows

rows = TermFrequencies(df1['names'])
df2 = pd.DataFrame(rows,columns = columns, index = list(df1['objectId']))
df2
```

```python
#maps-maps association matrix
#each entry corresponds to the dot product of the corresponding maps
c = df2.dot(df2.T)    

#Normalization
s = c.astype(float)
for i in c:
    for j in c:
        s[i][j] = c[i][j]/(c[i][i]+c[j][j]-c[i][j])

#final normalized association matrix
s
```

```python
#N Most similar maps
#data has to be normalized map-map association

def nsimilarMaps(df,s,n):

    top = []
    top1 = []
    index = []
    for m in s:
        index.append(df.loc[df['objectId'] == m, 'title'].iloc[0])
        x = s[m].nlargest(n+1)  #largest n scores for a map in association with other maps
        indxs = (x.index)
        maps = indxs[1:]      #exclude a map association with itself (score = 1)
        
        l=[]
        for i in maps:
            l.append(df.loc[df['objectId'] == i, 'title'].iloc[0])
        top.append([l])
        
        l1=[]
        for j in maps:
            l1.append( j )
        top1.append([l1])
    
    data = pd.DataFrame(top, index = index, columns = ['mostSimilar'])
    data1 = pd.DataFrame(top1, index = s.index, columns = ['mostSimilar'])
    return data,data1     

# 4 most similar maps using normalized dot product similarity
d, d1 = nsimilarMaps(df,s,4)
d
```

```python
#tf*idf

#These scores could be also imported since it takes time to calculate:    (file included)
#tfidf = pd.read_csv("tfidf.csv", index_col=0)


def tf_idf(dt):

    df = []
    tf = []
    idf = []
    N = len(dt.columns)
    for i in range(len(dt)):
        dfi = 0
        tfi = 0
        for t in range(N):
            cell = dt.iloc[i][t]
            if cell > 0:
                dfi += 1
                tfi += cell
        idfi = math.log((N/int(dfi)),2)
        idf.append(idfi)        
        df.append(dfi)
        tf.append(tfi) 
     
    data = pd.DataFrame(index = dt.index)
    data['df'] = df
    data['tf'] = tf
    data['idf'] = idf
    return data

tfidf = tf_idf(df2)
tfidf
```

```python
#tf*idf weights maps-terms matrix

#file weights is included and can be imported in the next block
def W(df2,tfidf)
    weights = []
    for r in range(len(df2)):
        weight = []
        for c in range(len(df2.columns)):
            tf = df2.iloc[r][c]
            idf = tfidf.iloc[r]['idf']
            w = tf*idf
            weight.append(w)
        weights.append(weight)
    weights = pd.DataFrame(weights, index = df2.index, columns = df2.columns)
    return weights
weights = W(df2,tfdif)
weights
```

```python
#This matrix is used to expand a query 
#######

#Euc are Euclidean distances of one term i with all other terms including itself -> (term i Euclidean distance vector)
from scipy.spatial.distance import squareform, pdist

Euc = pd.DataFrame(squareform(pdist(df2.T)), columns=df2.columns, index=df2.columns)
Euc
```

```python
#Expand query search
def expandQuery(Q,n,s1,sim):
    
    newQuery = []
    top = []
    for t in Q:
        newQuery.append(t)
        if t in s1.columns and sim == 'Euc':
            x = s1[t].nsmallest(n+1)  #smallest n scores in association with other terms
        elif t in s1.columns and sim == 'dot':
            x = s1[t].nlargest(n+1)  #largest n scores in association with other terms
        else:
            pass
        indxs = (x.index)
        expand = indxs[1:]      #exclude term association with itself (score = 1)
        l=[]
        for i in expand:
            l.append(i)
        top.append(l)
    
    length = 0
    for j in top:
        newQuery[length+1:length+1] = j
        length += len(j)+1

    return newQuery
    
Q0 = clear('engine united europe')

#data entry should be normalized using either 'Euc' or 'dot' similarity measures
Q1 = expandQuery(Q0,1,Euc,'Euc')
Q2 = expandQuery(Q0,1,dot_sim,'dot')

print(Q1,Q2)
```
